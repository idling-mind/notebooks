## Input file

Give the path to the file containing the data. The first column should be the angle in degrees and the next columns should be the values you want to plot. You can have one or more columns if you want and select only the columns you are interested in. The file should have a header row.

````python
input_file = 'polardata.csv'
````

## Columns to be plotted

````python
cols_to_plot = ['s1', 'seqv'] # The name of the column as per the header row
````

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
df = pd.read_csv(input_file) # Reading in the file
````

````python
df.head() # Previewing the file
````

````python
df.theta = df.theta * np.pi/180 # Converting angles to radians
````

````python
df.sort_values(by='theta', inplace=True) # Sorting the angles
````

````python
plt.figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k') # Setting image size and other properties
````

````python
ax = plt.subplot(111, projection='polar') # Creating the axes with polar projection
````

````python
# By default, matplotlib plots with zero on the east side and anti-clockwise direction.
# The following lines will rotate the plot by 90 degrees and also flip the direction.
ax.set_theta_offset(np.pi/2)
ax.set_theta_direction(-1)
````

````python
# Plot the quantities. If the name of your columns are different, set them appropriately here
for col in cols_to_plot:
    ax.plot(df[df.columns[0]], df[col])
````

````python
plt.show() # Show the plot
````
